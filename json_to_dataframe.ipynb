```python
TRAINING_PATH = "./Dataset/Training"
VALIDATION_PATH = "./Dataset/Validation"

Final_Training = pd.read_json(TRAINING_PATH + "/final_training.json")
Final_Validation = pd.read_json(VALIDATION_PATH + "/final_validation.json")
```

```python
# get persona-id from df_Final_Training['profile']
df_final_training = pd.DataFrame(Final_Training['profile'].apply(lambda x: x['persona-id']))
df_final_training.rename(columns={'profile': 'persona-id'}, inplace=True)
df_final_training
```

```python
df_final_training
```

```python
df_final_training['emotion-id'] = Final_Training['profile'].apply(lambda x : x['emotion']['emotion-id'])
df_final_training['type'] = Final_Training['profile'].apply(lambda x : x['emotion']['type'])
df_final_training['situation'] = Final_Training['profile'].apply(lambda x : x['emotion']['situation'])

df_final_training
```

```python
df_final_training['HS01'] = Final_Training['talk'].apply(lambda x : x['content']['HS01'])
df_final_training['SS01'] = Final_Training['talk'].apply(lambda x : x['content']['SS01'])
df_final_training['HS02'] = Final_Training['talk'].apply(lambda x : x['content']['HS02'])
df_final_training['SS02'] = Final_Training['talk'].apply(lambda x : x['content']['SS02'])
df_final_training['HS03'] = Final_Training['talk'].apply(lambda x : x['content']['HS03'])
df_final_training['SS03'] = Final_Training['talk'].apply(lambda x : x['content']['SS03'])

df_final_training
```

```python
df_final_validation = pd.DataFrame(Final_Validation['profile'].apply(lambda x: x['persona-id']))
df_final_validation.rename(columns={'profile': 'persona-id'}, inplace=True)
df_final_validation['emotion-id'] = Final_Validation['profile'].apply(lambda x : x['emotion']['emotion-id'])
df_final_validation['type'] = Final_Validation['profile'].apply(lambda x : x['emotion']['type'])
df_final_validation['situation'] = Final_Validation['profile'].apply(lambda x : x['emotion']['situation'])
df_final_validation['HS01'] = Final_Validation['talk'].apply(lambda x : x['content']['HS01'])
df_final_validation['SS01'] = Final_Validation['talk'].apply(lambda x : x['content']['SS01'])
df_final_validation['HS02'] = Final_Validation['talk'].apply(lambda x : x['content']['HS02'])
df_final_validation['SS02'] = Final_Validation['talk'].apply(lambda x : x['content']['SS02'])
df_final_validation['HS03'] = Final_Validation['talk'].apply(lambda x : x['content']['HS03'])
df_final_validation['SS03'] = Final_Validation['talk'].apply(lambda x : x['content']['SS03'])

df_final_validation
```

```python
df_final_validation.to_csv('./Dataset/df_final_validation.csv', index=False)
df_final_training.to_csv('./Dataset/df_final_training.csv', index=False)
```
